## 16.  set font and properties for title and labels

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randint(10,size=10)
y = np.random.randint(10, size=10)

font1 = {"family": "serif", "color": "red", "size": 20}
font2 = {"family": "serif", "color": "green", "size": 15}

plt.title("random value graph",fontdict=font1)
plt.xlabel('x-axis',fontdict=font2)
plt.ylabel("y-axis", fontdict=font2)

plt.plot(x,y)
plt.show()
```

```python
# multiple subplot

import matplotlib.pyplot as plt
import numpy as np

n_plot = 12

plt.figure(figsize=(10,12))  #to specify the size of fig

for i in range(1,n_plot+1):
    x = np.random.randint(10, size=5)
    y = np.random.randint(10, size=5)

    # print(n_plot // 3, 4, i)

    plt.subplot(n_plot//3,4,i)
    plt.plot(x,y)

plt.show()
```

## 22. subplot title

```python
# Title for each plot


import matplotlib.pyplot as plt
import numpy as np

n_plot = 6

plt.figure(figsize=(12, 12))  # to specify the size of fig

for i in range(1, n_plot + 1):
    x = np.random.randint(10, size=5)
    y = np.random.randint(10, size=5)

    # print(n_plot // 3, 4, i)

    plt.subplot(n_plot // 3, 3, i)
    plt.plot(x, y)
    plt.title(f"plot No: {i}")

plt.show()
```
